```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import networkx as nx
import random
import pandas as pd
```

```python
def generate_watts_strogatz_graph(n, k, p, seed=None):
    """
    Returns a randomised ring graph using the Watts-Strogatz model.

    :param n: the number of the nodes in the network
    :param k: the average degree of a node
    :param p: the rewiring probability between 0 and 1
    :param seed: seed for random generator.
    
    """
    if seed is not None:
        random.seed(seed)
    
    G = nx.empty_graph(n) # Creating an empty undirected graph with n nodes
    
    # Creating ring graph
    edges=list()
    for i in sorted(G.nodes()):
        edges += list(zip([i]*5, [(i+j)%n for j in range(1, k // 2 + 1)]))
    G.add_edges_from(edges)
    
    # Rewiring
    for source, target in edges:
        if random.random() < p:
            new_target = random.choice([x for x in G.nodes() if x not in [source]+G.neighbors(source)])
            G.remove_edge(source, target)
            G.add_edge(source, new_target)   
        
    return G, nx.average_shortest_path_length(G), nx.average_clustering(G)
```

```python
def watts_strogatz_sampling(n, k, p, num_trials=10):
    file_name = 'results/watts_strogatz_n_%s_k_%s_p_%s.csv' % (n, k, str(p).replace('.', '_'))
    G, l0, c0 = generate_watts_strogatz_graph(n, k, 0)
    with open(file_name, 'w+') as f:
        f.write('trial,l,l0,l_norm,c,c0,c_norm\n')
        for i in range(1, num_trials + 1):
            G, l, c = generate_watts_strogatz_graph(n, k, p)
            f.write('%s,%s,%s,%s,%s,%s,%s\n' % (i,l,l0,l/l0,c,c0,c/c0))
    df = pd.read_csv(file_name)
    return df.l_norm.mean(), df.l_norm.std(), df.c_norm.mean(), df.c_norm.std()
# nx.average_shortest_path_length(G), nx.average_clustering(G)
```

```python
def run_watts_strogatz_experiment(n, k, experiment_label='x', p_range=[0, 0.5, 1]):
    file_name = 'results/watts_strogatz_n_%s_k_%s_experiment_%s.csv' % (n, k, experiment_label)
    with open(file_name, 'w+') as f:
        f.write('p,l_mean,l_std,c_mean,c_std\n')
        for p in p_range:
            l_mean, l_std, c_mean ,c_std = watts_strogatz_sampling(n, k, p)
            f.write('%s,%s,%s,%s,%s\n' % (p, l_mean, l_std, c_mean ,c_std))
            
    df = pd.read_csv(file_name)
    plt.figure()
    ax = plt.plot(df.p, df.l_mean, 'b')
    plt.fill_between(df.p, df.l_mean-df.l_std, df.l_mean+df.l_std, color='b', alpha=0.2)

    plt.plot(df.p, df.c_mean, 'g')
    plt.fill_between(df.p, df.c_mean-df.c_std, df.c_mean+df.c_std, color='g', alpha=0.2)
    plt.xscale('log')

    plt.xlabel('p')
    plt.ylabel('L(p)/L(0) and C(p)/C(0)')
    plt.savefig('results/watts_strogatz_n_%s_k_%s_experiment_%s.pdf' % (n, k, experiment_label))
    
    return
```

```python
run_watts_strogatz_experiment(100, 20, 'a', [round(0.05*x, 4) for x in range(1, 21)])
```

```python
run_watts_strogatz_experiment(100, 20, 'b', [round(pow(0.63, x), 4) for x in range(1, 21)])
```
